# Visualisasi Data Klasifikasi Kopra

```python
# library untuk manipulasi data
import pandas as pd
import numpy as np

# library untuk visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Akuisisi Data

- Analisis Hasil Supervised Learning

```python
df_supervised_1 = pd.DataFrame({
    "K-NN": [0.717, 0.544, 0.799, 0.728, 0.837, 0.810, 0.842],
    "SVM": [0.701, 0.511, 0.788, 0.734, 0.853, 0.794, 0.842],
    "C45": [0.750, 0.560, 0.783, 0.745, 0.826, 0.810, 0.799],
    "GNB": [0.652, 0.457, 0.750, 0.685, 0.804, 0.739, 0.799],
    "LR": [0.669, 0.391, 0.777, 0.685, 0.821, 0.772, 0.821],
    "LDA": [0.723, 0.391, 0.750, 0.717, 0.799, 0.761, 0.804],
    "QDA" : [0.685, 0.424, 0.804, 0.728, 0.815, 0.788, 0.832]
}, index=['1', '2', '3', '4', '5', '6', '7']) * 100
```

```python
print(df_supervised_1)
```

```python
cmap = plt.get_cmap('RdYlGn', len(df_supervised_1.columns)-1)
```

```python
# Plotting
plt.figure(figsize=(15,5))

for i, column in enumerate(df_supervised_1.columns):
    plt.plot(df_supervised_1.index, df_supervised_1[column], marker='o', color=cmap(i), label=column)

# Menambahkan judul dan label sumbu
# plt.title('Results of classification of copra types using supervised learning \n with data train 90% and data test 10%', fontsize=12)
plt.ylabel('accuracy percentage', fontsize=10)
plt.xlabel('experiment', fontsize=10)

# Menampilkan legenda dan grafik
plt.ylim(ymin=0, ymax=100)
plt.legend(loc='lower left', ncol=3)
plt.grid(True)
plt.show()
```
