# Visualisasi Data Klasifikasi Kopra

```python
# library untuk manipulasi data
import pandas as pd
import numpy as np

# library untuk visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def line_chart_1(color, dataset):
    
    # set warna 
    cmap = plt.get_cmap(color, len(dataset.columns)-1)
    
    # Plotting
    plt.figure(figsize=(10,5))

    for i, column in enumerate(dataset.columns):
        plt.plot(dataset.index, dataset[column], marker='o', color=cmap(i), label=column, linewidth=2.5)
            
    # Menambahkan judul dan label sumbu
    # plt.title('Results of classification of copra types using supervised learning \n with data train 90% and data test 10%', fontsize=12)
    plt.ylabel('accuracy percentage', fontsize=10)
    plt.xlabel('experiment', fontsize=10)

    # Menampilkan legenda dan grafik
    plt.ylim(ymin=0, ymax=100)
    plt.legend(loc='lower left', ncol=7)
    plt.grid(True)
    plt.show()
```
